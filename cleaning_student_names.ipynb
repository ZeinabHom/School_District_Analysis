```python
import pandas as pd
import os
```

```python
student_data_to_load = os.path.join("resources","students_complete.csv")
```

```python
student_data_df = pd.read_csv(student_data_to_load)
```

```python
student_data_df.student_name
```

```python
student_name =student_data_df.student_name.tolist()
student_name
```

```python
student_to_fix = []
for i in student_name:
    if len(i.split())>2:
        student_to_fix.append(i)
len(student_to_fix)
```

```python
prefixes = []
for i in student_to_fix:
    prefixes.append(i.split()[0])
```

```python
set( prefixes)
prefixes = set(prefixes)
print(prefixes)
```

```python
for i in prefixes:
    if len(i)<=5:
        print(i)
        
```

```python
suffixes = []
for i in student_to_fix:
    suffixes.append(i.split()[-1])
print(suffixes)
```

```python
set(suffixes)
suffixes = set(suffixes)
print(suffixes)
```

```python
for i in suffixes:
    if len(i)<=4:
        print(i)
```

```python
prefixes_suffixes = ["Dr. ", "Mr. ","Ms. ", "Mrs. ", "Miss ", " MD", " DDS", " DVM", " PhD"]
for i in prefixes_suffixes:
    student_data_df.student_name=student_data_df.student_name.str.replace(i , "")
student_data_df
```
